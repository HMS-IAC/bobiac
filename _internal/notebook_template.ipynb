To use `ndv` and display the output we need to tag the cell containing `ndv.imshow(data)` with `remove-output` and then add in the immediate next cell `viewer.widget().children[1].snapshot()` only show a screenshot of the viewer. This cell must also be tagged with `remove-input` since is something that we don't want to see in the final rendered notebook.

> NOTE: you will see this effective only when building the notebook, not if you run it for example in VSCode.

An example is shown below.

```python
import numpy as np
import ndv
img = np.random.randint(0, 255, (10, 100, 100), dtype=np.uint8)
```

```python
viewer = ndv.imshow(img)
```

```python
# this is necessary to show only the image when the notebook is built
# otherwise, the viewer will show the a broken widget. Add the tag "remove-input" to
# completely hide this cell from the built notebook. Also add the tag "remove-output" to
# hide the previous cell to remove the broker output from that cell as well.
viewer.widget().children[1].snapshot()
```
